Q1: Ask the user for a string and check whether it is a palindrome or not.
        A palindrome is a string which is same when we read it forward & backward. Eg-“madam”, “racecar” etc.

```python
# Taking Str input
s = input('Enter String: ')
print('Input:- ',s)
if s==s[::-1]:
    print(f"{s} is a palindrome")
else:
    print(f"{s} is not a palindrome")
```

Q2: Given a list of integers compute the average of all numbers in the list.

```python
l = [2,7,5,10,9,3]
sum=0
for nums in l:
    sum+=nums

avg = sum/len(l)
print('The avg of nums in the List: ',avg)
```

Q3: Input two lists of integers from the user. Merge them into one list and sort the result.

```python
# Taking input of integers for two lists
l1 = list(map(int, input("Enter integers for the first list (space-separated): ").split()))
l2 = list(map(int, input("Enter integers for the second list (space-separated): ").split()))

merged_list = l1 + l2
print('Merged List: ',merged_list)

merged_list.sort()
print('Sorted List: ',merged_list)
```

Q4: Given a tuple of integers, create: 

    • A tuple of all even numbers 
    • A tuple of all odd numbers

```python
t = (1,2,3,4,5,6,7,8)
l1 = []
l2 = []
for i in t:
    if i%2==0:
        l1.append(i)
    else:
        l2.append(i)
t1 = tuple(l1)
t2 = tuple(l2)
print("A tuple of all even numbers: ",t1)
print("A tuple of all odd numbers: ",t2)
```

Q5: Create a dictionary where: • Keys = student names • Values = marks(integer)  

    Write a menu - based program where user presses a key(ʼAʼ,‘Bʼ,‘Cʼ,‘Dʼ)depending on the operation they want to perform on the dictionary:
    1. A - Add a student
    2. B - Update marks
    3. C - Search for a student
    4. D - Display all Students and marks


```python
students = {}

while True:
    choice = input("A-Add  B-Update  C-Search  D-Display  E-Exit : ")

    if choice == 'A':
        name = input("Name: ")
        students[name] = int(input("Marks: "))

    elif choice == 'B':
        name = input("Name: ")
        if name in students:
            students[name] = int(input("New Marks: "))
        else:
            print("Not found!")

    elif choice == 'C':
        name = input("Name: ")
        print(students.get(name, "Not found!"))

    elif choice == 'D':
        for n, m in students.items():
            print(n, ":", m)

    elif choice == 'E':
        break

    else:
        print("Invalid choice!")
```

Q6: Given a list of words: words = ["apple","banana","kiwi","cherry","mango"]
        
    Create a dictionary that maps each word to its length.
    Example:
    {"apple": 5, "banana": 6, "kiwi": 4, ...}

```python
# Given List
words = ["apple", "banana", "kiwi", "cherry", "mango"]
word_lengths = {}

for word in words:
    word_lengths[word] = len(word)

print(word_lengths)
```

Q7: Write a program that takes a string from the user and prints the number of spaces in the string.

```python
s = input('Enter string: ')
print('Input string - ',s)
spaces = 0
for ch in s:
    if ch==" ":
        spaces+=1

print(f'No. of Spaces: ',spaces)
```

Q8: Write a program to check whether two lists share no common elements.

    # share no common elements list1 =[1,2,3,4] list2 =[5,6,7,8]
    # share common elements list1 =[1,2,3] list2 =[3,4]

```python
l1 = [1,2,3,4]
l2 = [5,6]
s1 = set(l1)
s2 = set(l2)

if s1.intersection(s2):
    print('Two list share common elements')
else:
    print('Two list share no common elements')
```

Q9: Given a list, print all elements that appear more than once in the list.

```python
lst = [1, 2, 3, 2, 4, 5, 3, 6, 3]

duplicates = []

for item in lst:
    if lst.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Items that appears more than once in the list: ", duplicates)
```
